```python
'''subplot concept
Why use this approach / when it's useful

It helps you compare multiple related plots side by side (or stacked), without needing separate figures.

It keeps your visualization compact and organizes plots logically.

Useful when comparing different functions or datasets that share the same x-axis (or similar structure).
'''

x=np.linspace(1,10,70)
plt.figure()
plt.subplot(2,1,1)# 2=no of rows,1=no of columns,1=index from left to right
plt.plot(x,np.sin(x),"--")
plt.subplot(2,1,2)# 2nd index
plt.plot(x,np.cos(x),"--")
plt.show()
```

```python
sns.get_dataset_names()
df=sns.load_dataset("tips")
df
```

```python
# Relplot
# We use relplot() (from Seaborn) because it provides a simple, flexible, and powerful way to visualize relationships
# (i.e. correlations, trends, patterns) between variables in a dataset
plt.figure(figsize=(1,1))
sns.relplot(x="total_bill",y="tip",data=df.tail(35),hue="smoker",style="time",kind="scatter")
```

```python
# i want to know the bill generated for the each day in tips dataset
#catplot() is a figure-level function that draws plots showing the relationship-
#between a categorical variable (e.g. day of week, gender, species) and a numeric variable (e.g. score, salary, height).
# It lets you quickly create different types of categorical plots (bar, box, violin, strip, swarm, etc.) just by changing the kind parameter.
plt.figure(figsize=(2,2))
sns.catplot(x="day",y="total_bill",data=df,kind="violin",color="g")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1  Basic histogram of a numeric variable
data = sns.load_dataset("penguins")            # built-in dataset
sns.histplot(data=data, x="body_mass_g")       # histogram of body mass
plt.title("Distribution of Penguin Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example: sample city-population data (in thousands)
data = {
    "city": ["CityA","CityB","CityC","CityD","CityE","CityF","CityG","CityH","CityI","CityJ"],
    "population_thousands": [1200, 850, 4500, 3000, 1500, 2200, 950, 4100, 600, 1750]
}
df = pd.DataFrame(data)

# Plot histogram of population
sns.histplot(data=df, x="population_thousands", bins=5, kde=False, color="skyblue", edgecolor="black")
plt.xlabel("Population (thousands)")
plt.ylabel("Number of cities")
plt.title("Distribution of City Populations (sample data)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create some sample numeric data
data = np.random.normal(loc=0, scale=1, size=500)  # 500 random numbers from a normal distribution

# Plot histogram
sns.histplot(data=data, bins=20, color='skyblue', edgecolor='black')

# Add labels + title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Simple Histogram of Sample Data")

plt.show()
```
